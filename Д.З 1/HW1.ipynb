Обучение

```python
X=data[["NCD_sum","BL_sum","NAD_sum","AI_sum","NAC_sum","ND_sum","CS_sum","AT_sum","NA_sum","ADL_sum","AS(NA)_sum","AS(NAC)_sum"]]
Y=data[" predict"]
N, d = X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors = 15)
#knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)

y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

```

```python
from sklearn.metrics import mean_squared_error
err_train =mean_squared_error(y_train,y_train_predict)
err_test  = mean_squared_error(y_test,y_test_predict)
print(err_train, err_test)
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)

y_train_predict = svr.predict(X_train)
y_test_predict = svr.predict(X_test)

err_train =mean_squared_error(y_train,y_train_predict)
err_test  = mean_squared_error(y_test,y_test_predict)
print(err_train, err_test)
```

```python
from sklearn import ensemble
rf = ensemble.RandomForestRegressor(n_estimators = 1000)
rf.fit(X_train, y_train)

y_train_predict = rf.predict(X_train)
y_test_predict = rf.predict(X_test)

err_train =mean_squared_error(y_train,y_train_predict)
err_test  = mean_squared_error(y_test,y_test_predict)
print(err_train, err_test)
```

```python
ert = ensemble.ExtraTreesRegressor(n_estimators = 100).fit(X_train, y_train)

y_train_predict = ert.predict(X_train)
y_test_predict = ert.predict(X_test)

err_train =mean_squared_error(y_train,y_train_predict)
err_test  = mean_squared_error(y_test,y_test_predict)
print(err_train, err_test)
```

```python
ada = ensemble.AdaBoostRegressor(n_estimators = 1000)
ada.fit(X_train, y_train)

y_train_predict = ada.predict(X_train)
y_test_predict = ada.predict(X_test)

err_train =mean_squared_error(y_train,y_train_predict)
err_test  = mean_squared_error(y_test,y_test_predict)

print(err_train, err_test)
```

```python
gbt = ensemble.GradientBoostingRegressor(n_estimators = 1000)
gbt.fit(X_train, y_train)

y_train_predict = gbt.predict(X_train)
y_test_predict = gbt.predict(X_test)

err_train =mean_squared_error(y_train,y_train_predict)
err_test  = mean_squared_error(y_test,y_test_predict)

print(err_train, err_test)
```
